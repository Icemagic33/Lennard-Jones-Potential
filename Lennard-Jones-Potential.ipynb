# Lennard-Jones Potential and Simple Harmonic Motion

## Introduction
The Lennard-Jones potential is a mathematical model that describes the interaction between a pair of neutral atoms or molecules based on their distance. This potential is particularly useful in simulating the behavior of gases and liquids in molecular dynamics.

## Lennard-Jones Potential
The Lennard-Jones potential \( V(r) \) is expressed as:
$$ V(r) = 4\epsilon \left[ \left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6 \right] $$
where:
- $( r )$ is the distance between particles,
  
- $( \epsilon )$ is the depth of the potential well, indicating the strength of the attraction,
  
- $( \sigma )$ is the distance at which the potential energy is zero, representing the effective diameter of the molecules.


## Equilibrium Position
The equilibrium position $( r_{\text{eq}} )$ is where the force $( F(r) )$ derived from the potential is zero. The force is the negative gradient of the potential:
$$ F(r) = -\frac{dV}{dr} = -4\epsilon \left[ 12\left(\frac{\sigma}{r}\right)^{13} - 6\left(\frac{\sigma}{r}\right)^7 \right] $$
Setting $( F(r) = 0 )$ finds the equilibrium typically at $( r = 2^{1/6}\sigma )$.


## Taylor Expansion Around Equilibrium
For small oscillations around the equilibrium, we perform a Taylor expansion:
$$V(r) \approx V(r_{\text{eq}}) + \frac{1}{2} V''(r_{\text{eq}})(r - r_{\text{eq}})^2 $$
where \$( V''(r_{\text{eq}}) \)$ is the second derivative of \( V \) at the equilibrium.


## Simple Harmonic Motion
The effective spring constant \( k \) in SHM is derived from the curvature of the potential well at the equilibrium:
$$ k = V''(r_{\text{eq}}) = 4\epsilon \left[ 156\left(\frac{\sigma}{2^{1/6}\sigma}\right)^{14} - 42\left(\frac{\sigma}{2^{1/6}\sigma}\right)^8 \right] $$
This leads to SHM for small displacements \$( x = r - r_{\text{eq}} \)$:
$$ m\ddot{x} + kx = 0 $$
indicating the motion is simple harmonic, with restoring force proportional to displacement.


## Energy Conservation in SHM
In SHM, the total energy \( E \) is constant and given by:
$$ E = \frac{1}{2} m \dot{x}^2 + \frac{1}{2} kx^2 $$
where \$( \dot{x} \)$ is the velocity, and the terms represent kinetic and potential energy respectively.


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the Lennard-Jones potential
epsilon = 1.0  # Depth of the potential well
sigma = 1.0    # Finite distance at which the inter-particle potential is zero

# Define the Lennard-Jones potential function
def lennard_jones_potential(r):
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

# Define the force function derived from the Lennard-Jones potential
def lj_force(r):
    return 24 * epsilon * (2 * (sigma / r)**13 - (sigma / r)**7) / r

# Define Taylor expansion around the equilibrium point r_eq
def taylor_expansion(r, r_eq, V_eq, V_second):
    return V_eq + 0.5 * V_second * (r - r_eq)**2

# Calculate values for plotting
r = np.linspace(0.95 * sigma, 3 * sigma, 400)
V = lennard_jones_potential(r)
F = lj_force(r)

# Equilibrium position (r_eq) is approximately at the minimum of the potential
r_eq = r[np.argmin(V)]
V_eq = lennard_jones_potential(r_eq)
V_second = (lennard_jones_potential(r_eq + 0.01) - 2 * V_eq + lennard_jones_potential(r_eq - 0.01)) / 0.01**2

# Taylor expansion around r_eq
V_taylor = taylor_expansion(r, r_eq, V_eq, V_second)

# Plotting
plt.figure(figsize=(12, 6))

# Plot the Lennard-Jones potential
plt.subplot(1, 2, 1)
plt.plot(r, V, label='Lennard-Jones Potential')
plt.plot(r, V_taylor, 'r--', label='Taylor Expansion around $r_{eq}$')
plt.xlabel('Distance (r)')
plt.ylabel('Potential Energy (V)')
plt.legend()
plt.title('Lennard-Jones Potential and Taylor Expansion')
plt.grid(True)

# Plot the force
plt.subplot(1, 2, 2)
plt.plot(r, F, label='Force (-dV/dr)')
plt.xlabel('Distance (r)')
plt.ylabel('Force (F)')
plt.legend()
plt.title('Force derived from Lennard-Jones Potential')
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Analysis of the Lennard-Jones Potential and Force

## Lennard-Jones Potential and Taylor Expansion

The plot on the left displays the Lennard-Jones potential $( V(r) )$ against the distance $( r )$ between two particles. The blue line represents the actual potential energy calculated using the Lennard-Jones formula:
\$[ V(r) = 4\epsilon \left[ \left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6 \right] \]$

where $( \epsilon )$ represents the depth of the potential well and $( \sigma )$ is the distance at which the inter-particle potential is zero.

The red dashed line shows the Taylor expansion around the equilibrium position $( r_{\text{eq}} )$, which is approximately at $( r = 2^{1/6} \sigma )$. This expansion is used to approximate the potential near the equilibrium point, showing how the potential can be viewed as a quadratic function, characteristic of simple harmonic motion for small deviations from equilibrium. This provides an excellent approximation close to $( r_{\text{eq}} )$ but diverges at greater distances where the non-linear nature of the Lennard-Jones potential becomes significant.


## Force Derived from the Lennard-Jones Potential

The plot on the right illustrates the force derived from the gradient of the Lennard-Jones potential:
$[ F(r) = -\frac{dV}{dr} = -4\epsilon \left[ 12\left(\frac{\sigma}{r}\right)^{13} - 6\left(\frac{\sigma}{r}\right)^7 \right] ]$
This graph shows the force as a function of distance $( r )$. The force is attractive (negative) beyond the equilibrium distance, pulling particles together when they are too far apart, and repulsive (positive) below this distance, pushing them apart when they are too close. The force crosses zero at the equilibrium distance, indicating no net force on the particles  a necessary condition for stable equilibrium.
